## Question 2

Explain the relationship between **Primary Keys** and **Foreign Keys**.

Type your answer below this line:
 

A primary key - foreign key relationship defines a one-to-many relationship between two tables in a relational database. A foreign key is a column or a set of columns in one table that references the primary key columns in another table. The primary key is defined as a column (or set of columns) where each value is unique and identifies a single row of the table.

## Question 4

Explain the various types of JOINs possible with SQL. 

Type your answer below this line:
Inner Join : The common elements, meaning displays rows that are present in both tables.

Left Join: Show all the rows in the first/left table and the common rows with the second table.

Right Join: Get all the rows in the second/right table and the common rows of the first table.

Outer Join: Get all rows in both tables.
